```python
Question 2:- A dataset exported from a database contains the following customer age values.

ages = [25, 42, -1, 34, None, 52, 17, -5, 29, None, 46] Some entries are invalid because they are either negative numbers or missing values (None).

Write a Python program that:

Iterates through the dataset and removes all invalid age values.
Stores only valid ages in a new list.
Calculates the average age of valid customers.
Displays the list of customers whose age is greater than 30.
Prints how many invalid entries were removed during the cleaning process.
```

```python
#Iterates through the dataset and removes all invalid age values.

ages = [25, 42, -1, 34, None, 52, 17, -5, 29, None, 46] 

invalid_count = 0
for age in ages:
    if age is None or age < 1:
        ages.remove(age)
        invalid_count = invalid_count + 1

#Prints how many invalid entries were removed during the cleaning process.
print(invalid_count)
        

#Stores only valid ages in a new list.        
new_ages = ages        
print(new_ages)

# Calculates the average age of valid customers.

sum = 0
average = 0
for age in new_ages:
    sum = sum + age
    
average = sum / len(new_ages)
print(average)

#Displays the list of customers whose age is greater than 30.
above30_age = []
for age in new_ages:
    if age > 30:
      above30_age.append(age)  
print(above30_age)
        

```

```python
login_hours = [1, 3, 5, 7, 2, 6, 8, 4, 9, 10]

# 1. Separates login hours into even-hour logins and odd-hour logins.
even_hours = []
odd_hours = []
for hour in login_hours:
    if hour % 2 == 0:
        even_hours.append(hour)
    elif hour % 2 != 0:
        odd_hours.append(hour)
    else:
        print("No even or odd hours found")
print(even_hours)
print(odd_hours)

# 2. Identifies login hours that fall into peak activity time (hours greater than 5).
peak_hours = []
count_peak = 0
for hour in login_hours:
    if hour > 5:
        peak_hours.append(hour)
        count_peak = count_peak + 1

# 4. Calculates how many logins occurred during peak hours.        
print(count_peak )



#3. Sorts the login hours in ascending order.
login_hours.sort()
print(login_hours)
```

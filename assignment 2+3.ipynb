Q1. Division with Exception Handling

Write a program that takes two numbers and safely performs division.
Handle division by zero using try-except.

```python
x = int(input("Enter a number:"))
y = int(input("Enter a number:"))

def division(x, y):
    try:  
        return x/y
        
    except Exception as e:
        print(e)
        
print(division(x, y))
```

Q2. String to Integer Conversion

Given a list:

data = ["10", "20", "abc", "30"]

Convert all values to integers.
Skip invalid values using exception handling.

```python
data = ["10", "20", "abc", "30"]
int_data = []

for i in data:
    try:
        if type(int(i)) == int:
            int_data.append(int(i))
    except ValueError:
        print("Not an integer")
   
print(int_data)

```

Q3. Multiple Exception Handling

Write a program that:

Takes user input
Converts it to integer
Divides 100 by that number

Handle:

ValueError
ZeroDivisionError

```python
num = int(input("Enter a number:"))
try:
    100/num
except ValueError:
    print("Value Error")
except ZeroDivisionError:
    print("Zero Division Error")
```

```python

try:
    file = open("data.txt")
except:
    print("File not found")
finally:
    file.close()
 
```
